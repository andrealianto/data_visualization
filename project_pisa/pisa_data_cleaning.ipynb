# PISA 2012 Data Cleaning

```python
import pandas as pd
import numpy as np

%pylab inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = 'pisa2012.csv'

# load csv file
pisa = pd.read_csv(filename)
```

## 1. Create score table

```python
# Create score table
score_columns = ['CNT', 'OECD',
                 'PV1MATH', 'PV2MATH', 'PV3MATH', 'PV4MATH', 'PV5MATH',
                 'PV1READ', 'PV2READ', 'PV3READ', 'PV4READ', 'PV5READ',
                 'PV1SCIE', 'PV2SCIE', 'PV3SCIE', 'PV4SCIE', 'PV5SCIE',
                 'W_FSTUWT']

# subset original dataframe
pisaScores = pisa[score_columns].copy()
pisaScores = pisaScores.fillna(0)
```

```python
pisaScores.head()
```

```python
# Function to aggregate math, reading, and science scores by weighted average calculation
def aggregate_score(df):
    df['PV1MATH_temp'] = df.PV1MATH * df.W_FSTUWT
    df['PV2MATH_temp'] = df.PV2MATH * df.W_FSTUWT
    df['PV3MATH_temp'] = df.PV3MATH * df.W_FSTUWT
    df['PV4MATH_temp'] = df.PV4MATH * df.W_FSTUWT
    df['PV5MATH_temp'] = df.PV5MATH * df.W_FSTUWT
    
    df['PV1READ_temp'] = df.PV1READ * df.W_FSTUWT
    df['PV2READ_temp'] = df.PV2READ * df.W_FSTUWT
    df['PV3READ_temp'] = df.PV3READ * df.W_FSTUWT
    df['PV4READ_temp'] = df.PV4READ * df.W_FSTUWT
    df['PV5READ_temp'] = df.PV5READ * df.W_FSTUWT

    df['PV1SCIE_temp'] = df.PV1SCIE * df.W_FSTUWT
    df['PV2SCIE_temp'] = df.PV2SCIE * df.W_FSTUWT
    df['PV3SCIE_temp'] = df.PV3SCIE * df.W_FSTUWT
    df['PV4SCIE_temp'] = df.PV4SCIE * df.W_FSTUWT
    df['PV5SCIE_temp'] = df.PV5SCIE * df.W_FSTUWT
    
    grouped = df.groupby(['CNT', 'OECD']).sum().reset_index()
    grouped.head()
    
    grouped['PV1MATH_AVE'] = grouped.PV1MATH_temp / grouped.W_FSTUWT
    grouped['PV2MATH_AVE'] = grouped.PV2MATH_temp / grouped.W_FSTUWT
    grouped['PV3MATH_AVE'] = grouped.PV3MATH_temp / grouped.W_FSTUWT
    grouped['PV4MATH_AVE'] = grouped.PV4MATH_temp / grouped.W_FSTUWT
    grouped['PV5MATH_AVE'] = grouped.PV5MATH_temp / grouped.W_FSTUWT
    grouped['MATHAVE'] = (grouped.PV1MATH_AVE + grouped.PV2MATH_AVE + grouped.PV3MATH_AVE + 
                          grouped.PV4MATH_AVE + grouped.PV5MATH_AVE)/5
    
    grouped['PV1READ_AVE'] = grouped.PV1READ_temp / grouped.W_FSTUWT
    grouped['PV2READ_AVE'] = grouped.PV2READ_temp / grouped.W_FSTUWT
    grouped['PV3READ_AVE'] = grouped.PV3READ_temp / grouped.W_FSTUWT
    grouped['PV4READ_AVE'] = grouped.PV4READ_temp / grouped.W_FSTUWT
    grouped['PV5READ_AVE'] = grouped.PV5READ_temp / grouped.W_FSTUWT
    grouped['READAVE'] = (grouped.PV1READ_AVE + grouped.PV2READ_AVE + grouped.PV3READ_AVE + 
                          grouped.PV4READ_AVE + grouped.PV5READ_AVE)/5
    
    grouped['PV1SCIE_AVE'] = grouped.PV1SCIE_temp / grouped.W_FSTUWT
    grouped['PV2SCIE_AVE'] = grouped.PV2SCIE_temp / grouped.W_FSTUWT
    grouped['PV3SCIE_AVE'] = grouped.PV3SCIE_temp / grouped.W_FSTUWT
    grouped['PV4SCIE_AVE'] = grouped.PV4SCIE_temp / grouped.W_FSTUWT
    grouped['PV5SCIE_AVE'] = grouped.PV5SCIE_temp / grouped.W_FSTUWT
    grouped['SCIEAVE'] = (grouped.PV1SCIE_AVE + grouped.PV2SCIE_AVE + grouped.PV3SCIE_AVE + 
                          grouped.PV4SCIE_AVE + grouped.PV5SCIE_AVE)/5
    
    return grouped
    
grouped = aggregate_score(pisaScores)

# Drop unnecessary columns
wanted_columns = ['CNT', 'OECD', 'MATHAVE', 'READAVE', 'SCIEAVE']
pisaScores_grouped = grouped[wanted_columns].copy()
pisaScores_grouped.head()
```

## 2. Create table containing other supporting variables

```python
# Choose variables
wanted_columns = ['CNT',                       # Country
                  'PARED',                     # Highest parental education in years
                  'LMINS', 'MMINS', 'SMINS',   # Learning time for language, maths, and science in school (in mins)
                  'OUTHOURS',                  # Out-of-school study time (in hours)
                  'ST57Q01', 'ST57Q02', 'ST57Q03', 'ST57Q04', 'ST57Q05', 'ST57Q06', # How out-of-school study time was spent
                  'TIMEINT'                    # Time of computer use (in mins)
                 ] 

# subset original dataframe
pisaSupport = pisa[wanted_columns].copy()
pisaSupport = pisaSupport.fillna(0)

# renaming columns
pisaSupport=pisaSupport.rename(columns = {'ST57Q01':'OUT_homework', 'ST57Q02':'OUT_guided_homework',
                                          'ST57Q03':'OUT_personal_tutor', 'ST57Q04': 'OUT_commercial_company',
                                          'ST57Q05':'OUT_parents', 'ST57Q06':'OUT_computer'})

# Convert variables from minutes into hours
pisaSupport['LHOURS'] = pisaSupport['LMINS'] / 60
pisaSupport['MHOURS'] = pisaSupport['MMINS'] / 60
pisaSupport['SHOURS'] = pisaSupport['SMINS'] / 60
pisaSupport['SCHHOURS'] = pisaSupport['LHOURS'] + pisaSupport['MHOURS'] + pisaSupport['SHOURS']
pisaSupport['COMHOURS'] = pisaSupport['TIMEINT'] /60

pisaSupport_grouped = pisaSupport.groupby('CNT').mean().reset_index()
pisaSupport_grouped.head()
```

## 3. Merge score and out-of-school-study tables

```python
# Merge score table with out-of-school-study table
pisaMerged = pd.merge(pisaScores_grouped, pisaSupport_grouped, how = 'left', on = 'CNT')
pisaMerged.head()
```

## 4. Drop non-country rows

```python
# Check number of countries to see if there is any duplication
len(pisaMerged)
```

According to the dataset description, only 65 countries participated in PISA 2012.

A quick check of the CSV file revealed that in the dataset, USA is split into 3 regions, and Russsian Federation is split into 2 regions. We will aggregate Connecticut, Florida, and Massachusetts into USA. We will also aggregate Perm(Russian Federation) and Russian Federation into Russia.

```python
# Drop non-country rows (Connecticut, Massachusetts, Florida, Perm)
idxUSA = pisaMerged[pisaMerged['CNT'].str.contains("USA")].index.tolist()
idxRussia = pisaMerged[pisaMerged['CNT'].str.contains("Perm")].index.tolist()
dropped_idx = idxUSA + idxRussia
pisaMerged = pisaMerged.drop(dropped_idx)
len(pisaMerged)
```

## 5. Export to CSV file

```python
# output csv file
pisaMerged.to_csv('pisa2012_final.csv',  index=False)
```
